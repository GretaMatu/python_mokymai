```python
import pandas as pd
```

```python
salaries = pd.read_csv('salary_data.csv')
salaries
```

```python
salaries.plot.scatter(x='YearsExperience', y='Salary')
```

```python
from sklearn import linear_model
```

```python
x = salaries[['YearsExperience']].to_numpy()
x.shape
```

```python
Y = salaries[['Salary']].to_numpy()
Y.shape
```

```python
model = linear_model.LinearRegression()
model
```

```python
model.predict(5) # nera apmokytas dar
```

```python
model.fit(x,Y)
```

```python
model.predict( [[5], [10]] )
```

```python
predictions = model.predict(x)
predictions
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, Y)
plt.plot( x, predictions, c='r')
```
